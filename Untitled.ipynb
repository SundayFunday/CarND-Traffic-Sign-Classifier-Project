```python
%matplotlib inline
import numpy as np
from six.moves import cPickle as pickle
import matplotlib.pylab as plt
from matplotlib.gridspec import GridSpec
import cv2
```

```python
with open('test.p', 'rb') as test_file:
    test_dict = pickle.load(test_file)
X_test,y_test = test_dict['features'], test_dict['labels']
```

```python
def group(labels):
    unique, counts = np.unique(labels, return_counts=True)
    return np.asarray((unique, counts)).T
```

```python
def get_label_map(f_name, labels):
    mapper = np.genfromtxt(f_name, delimiter=',', usecols=(1,), dtype=str, skip_header=1)
    crapper = np.vectorize(lambda x : mapper[x])
    return crapper(labels)
```

```python
def plot_class_frequencies(labels):
    freqs = group(labels)
    plt.figure(figsize=(15,5))
    plt.bar(freqs[:,0], freqs[:,1])
    plt.xlabel('ClassID')
    plt.ylabel('Frequency')
    ind = np.arange(0.5,43.5)
    plt.xticks(ind, get_label_map('signnames.csv', np.unique(y_test)),  ha='right', rotation=45)
    plt.show()
```

```python
plot_class_frequencies(y_test)
```

```python
def get_unique_images(images, labels):
    class_id, idx = np.unique(labels, return_index=True)
    return images[idx], class_id
```

```python
images, labels = get_unique_images(X_test, y_test)
images.shape
```

```python
def plot_images(images, labels):
    gs = GridSpec(6, 7)
    gs.update(wspace=0.03, hspace=0.03) # set the spacing between axes. 
    fig = plt.figure(figsize=(12,12))
    image_titles = get_label_map('signnames.csv', labels)
    for i in range(len(images)-1):
        ax = fig.add_subplot(gs[i])
        img = images[i]
        ax.imshow(img)
        ax.set_aspect('equal')
        ax.set_title(image_titles[i],fontsize=7)
        plt.axis('off')
    gs.tight_layout(fig)
    plt.show()
```

```python
plot_images(images, labels)
```
